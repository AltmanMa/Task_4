```python
codeList = df['Code'].tolist()
MarketList = []
for stock in codeList:
    Marketdata = []
    data = api.get_eod_historical_stock_market_data(f"{stock}.US", from_date= '2019-06-28', to_date= '2024-06-30', period = 'w')
    for daily in data:
        date = daily['date']
        volume = daily['volume']
        try:
            cap = api.get_historical_market_capitalization_data(stock, from_date= date, to_date = date)['0']['value']
        except Exception as e:
            print(f"{stock} cap error on {date}")
            cap = 0
        Marketdata.append([date, volume, cap])
    MarketList.append([stock, Marketdata])
```
